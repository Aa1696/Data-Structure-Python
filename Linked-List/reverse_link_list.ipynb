```python
class Node:
    def __init__(self,data=None,next=None):
        self.data=data
        self.next=next
        
class Link_List:
    def __init__(self):
        self.head=None
        
    def reverse(self):
        prev=None
        curr=self.head
        while(curr):
            next=curr.next
            curr.next=prev
            prev=curr
            curr=next
        self.head=prev
        
    def insert_at_begning(self,data):
        node=Node(data,self.head)
        self.head=node
        
    def insert_at_end(self,data):
        itr=self.head
        while(itr):
            if(itr.next is None):
                node=Node(data,itr.next)
#                 node.next=itr.next
                itr.next=node
                break
            itr=itr.next
    def insert_at_index(self,data,index):
        if(index<0 or index>self.get_length()):
            raise Exception("There is an error")
        itr=self.head
        count =0
        while(itr):
            if(count==index-2):
                node=Node(data,itr.next)
                itr.next=node
                break
            itr=itr.next
            count +=1
            
    def get_length(self):
        if(self.head is None):
            return 0
        count=0
        itr=self.head
        while(itr):
            count +=1
            itr=itr.next
        return count
        
        
    def print_lnklist(self):
        if(self.head is None):
            print("Linked_List is empty")
            return
        itr=self.head
        tmp=''
        while(itr):
            tmp +=str(itr.data)+'-->'
            itr =itr.next
        print(tmp,self.head)
```

```python
lnk=Link_List()
print("Insert at begning")
lnk.insert_at_begning(2)
lnk.insert_at_begning(4)
lnk.insert_at_begning(5)
lnk.insert_at_begning(6)
lnk.insert_at_begning(8)
lnk.insert_at_index(12,3)
lnk.print_lnklist()
print("Length of Link_List :",lnk.get_length())
print("Insert at end")
lnk.insert_at_end(10)
lnk.insert_at_end(11)
lnk.insert_at_end(12)
lnk.insert_at_end(13)
lnk.insert_at_end(14)
lnk.print_lnklist()
print("Reverse the Link-List")
lnk.reverse()
lnk.print_lnklist()
print("Length of Link_List :",lnk.get_length())
```
