```python
class Node:
    def __init__(self,data=None,next=None):
        self.data=data
        self.next=next
        
            
class Link_List:
    def __init__(self):
        self.head=None
        
        
    def insert_at_begning(self,data):
        node=Node(data,self.head)
        self.head=node
        
    def delete_at_begning(self):
        if(self.head.next==None):
            self.head=None
            return
        self.head=self.head.next
    def delete_at_end(self):
        if(self.head.next==None):
            self.head=None
            return
        itr=self.head
        while(itr):
            if(itr.next.next==None):
                itr.next=None
                break
                
            itr=itr.next
            
    def delete_at_any_index(self,index):
        if(self.head.next==None):
            self.head=None
            return
        
        count=0
        itr=self.head
        while(itr):
            if(count==index-2):
                itr.next=itr.next.next
                break
            itr=itr.next
            count +=1
        
        
        
    def print_linklist(self):
        if(self.head is None):
            print("Link-List is Empty")
            return
        itr=self.head
        tmp=''
        while(itr):
            tmp +=str(itr.data) + '-->'
            itr=itr.next
        print(tmp)
        
        
    
            
        
        
```

```python
lnk=Link_List()
lnk.insert_at_begning(12)
lnk.insert_at_begning(13)
lnk.insert_at_begning(14)
lnk.insert_at_begning(15)
lnk.insert_at_begning(16)
lnk.insert_at_begning(17)
lnk.insert_at_begning(18)
lnk.insert_at_begning(19)
lnk.insert_at_begning(20)
lnk.insert_at_begning(21)
lnk.print_linklist()
print("Delete the first element")
lnk.delete_at_begning()
print("Print the Link_List")
lnk.print_linklist()
print("Delete the last element")
lnk.delete_at_end()
print("Print the Link_List")
lnk.print_linklist()
print("Delete element at given index")
lnk.delete_at_any_index(3)
lnk.print_linklist()
```
