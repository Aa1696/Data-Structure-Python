```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
        self.prev=None
        
        
        
class Double_link_list:
    def __init__(self):
        self.head=None
        
    def insertion_at_begning(self,data):
        tmp=self.head
        node=Node(data)
        node.next=self.head
        if(self.head is not None):
            self.head.prev=node
        self.head=node
        
    def insertion_at_end(self,data):
        itr=self.head
        if self.head is None:
            self.head = Node(data)
            return
        while(itr.next):
            lst=itr.next
            itr=itr.next
        node=Node(data)
        lst.next=node
        node.prev=lst
    def insert_at(self,data,index):
        if(index==0):
            self.insertion_at_begning(data)
            return
        tmp=self.head
        node=Node(data)
        for i in range(index-1):
            if(tmp !=None):
                tmp=tmp.next
        if(tmp.next != None):
            node.next=tmp.next
            node.prev=tmp
            tmp.next=node
            if(node.next is not None):
                node.next.prev=node
    def get_length(self):
        count=0
        itr=self.head
        while(itr):
            count +=1
            itr=itr.next
        return count
    
    
    def print_dbl_lnk_lst(self,node):
#         itr=self.head
        tmp_rt=''
        tmp_lft=''
        print("Read in a forward directio:")
        while(node):
            tmp_rt +=str(node.data) +'-->'
            last=node
            node=node.next
        print(tmp_rt)
        print("Read in a backward directio:")
        while(last):
            tmp_lft +=str(last.data) +'-->'
            last=last.prev
        print(tmp_lft)
```
