```python
class Node:
    def __init__(self,data=None,next=None):
        self.data=data
        self.next=next

class Linked_List:
    def __init__(self):
        self.head=None
        
    def insert_at_begning(self,data):
        node=Node(data,self.head)
        self.head=node
        
    def insert_at_end(self,data):
        if(self.head is None):
            self.insert_at_begning(data)
            return 

        tmp=self.head
        while(tmp):
            if(tmp.next is None):
                node=Node(data,None)
                tmp.next=node
            tmp=tmp.next
    def insert_at_given_index(self,data,index):
        if(index<0 or index>self.get_to_length()):
            raise Exception('There is a error')
        if(index==0):
            self.insert_at_begning(data)
            return
        count =0
        tmp=self.head
        while(tmp):
            if(count==index-2):
                node=Node(data,tmp.next)
                tmp.next=node
                break
            tmp=tmp.next
            count +=1
            
    def insert_after_value(self,data_after,data_to_insert):
        # Search for first occurance of data_after value in linked list
        # Now insert data_to_insert after data_after node
        
        if(self.head is None):
            return
        if(self.head.data==data_after):
            self.head.next=Node(data_to_insert,self.head.next)
            return
        itr=self.head
        while(itr):
            if(itr.data==data_after):
                itr.next=Node(data_to_insert,itr.next)
                break
            itr=itr.next
            
    def remove_by_value(self, data):
    # Remove first node that contains data i.e delete node just after the node that contains data
        if(self.head is None):
            return
        if(self.head.data==data):
            self.head=self.head.next
            return
        itr=self.head
        while(itr):
            if(itr.data==data):
                itr.next=itr.next.next
                break
            itr=itr.next
    
                
    def get_to_length(self):
        count=0
        tmp=self.head
        while(tmp):
            count +=1
            tmp=tmp.next
        return count
    def print(self):
        if(self.head is None):
            print("Linked list is empty")
            return
        tmp=self.head
        lst=''
        while(tmp):
            lst += str(tmp.data) +'-->'
            tmp =tmp.next
        print(lst)
        
```

```python
lnk=Linked_List()
lnk.insert_at_begning(13)
lnk.insert_at_begning(18)
lnk.insert_at_given_index(3,2)
lnk.insert_after_value(13,34)
# lnk.remove_by_value(34)
print(lnk.get_to_length())
lnk.print()
```
