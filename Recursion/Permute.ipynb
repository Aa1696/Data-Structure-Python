```python
def permute(s,rest, answer):
    if (len(s) == 0):
        answer.append(rest)
        return
      
    for i in range(len(s)):
        ch = s[i]
        left_substr = s[0:i]
        right_substr = s[i + 1:]
        word = left_substr + right_substr
        permute(word, rest+[ch],ans)
```

```python
s=["A","B","C"]
ans=[]
nw_ans=[]
permute(s,nw_ans,ans)
```

```python
from collections import defaultdict
dict1=defaultdict(list)
a=[2,3,4]
for i in range(2,10):
    if(i==2):
        dict1[i]=["A","B","C"]
    if(i==3):
        dict1[i]=["D","E","F"]
    if(i==4):
        dict1[i]=["G","H","I"]
    if(i==5):
        dict1[i]=["J","K","L"]
    if(i==6):
        dict1[i]=["M","N","O"]
    if(i==7):
        dict1[i]=["P","Q","R","S"]
    if(i==8):
        dict1[i]=["T","U","V"]
    if(i==9):
        dict1[i]=["W","X","Y","Z"]
new_str=[]
ans=[]
nw_lst=[]
for i in a:
    new_str +=dict1[i]
```

```python
def comb(new_str,nw_lst,ans):
    if(len(new_str)==0):
        ans.append(nw_lst)
        return 
    else:
        for i in range(len(new_str)):
            ch=new_str[i]
            left=new_str[:i]
            right=new_str[i+1:]
            rest=left+right
            comb(rest,nw_lst+[ch],ans)
```

```python
comb(new_str,nw_lst,ans)
```
