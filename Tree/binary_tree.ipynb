```python
class Binarysearchtree:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
        
    def add_child(self,data):
        if(self.data==data):
            return 
        
        if(self.data):
            if(data>self.data):
                if(self.right):
                    self.right.add_child(data)
                else:
                    self.right=Binarysearchtree(data)
            else:
                if(self.left):
                    self.left.add_child(data)
                else:
                    self.left=Binarysearchtree(data)
        else:
            self.data=data
    def height(self,node):
        if(node is None):
            return -1
        left_height=self.height(node.left)
        right_height=self.height(node.right)
        return 1 +max(left_height,right_height)
    def size(self):
        size=1
        if(self.left):
            size +=self.left.size()
        if(self.right):
            size +=self.right.size()
        return size
    def inorder_traversal(self):
        ele=[]
        #In inorder first left then root then right
        #visiting left
        if(self.left):
            ele +=self.left.inorder_traversal()
        
        #visiting root after traversing left sub tree
        ele.append(self.data)
        
        #visiting right subtree
        if(self.right):
            ele +=self.right.inorder_traversal()
            
        return ele
    def preorder_traversal(self):
        ele=[]
        #visit root
        ele.append(self.data)
        
        #visit left_subtree
        if(self.left):
            ele +=self.left.preorder_traversal()
        #visit right_subtree
        if(self.right):
            ele +=self.right.preorder_traversal()
        return ele
    def postorder_traversal(self):
        ele=[]
        
        #visiting left
        if(self.left):
            ele +=self.left.postorder_traversal()
            
        #visiting right
        if(self.right):
            ele +=self.right.postorder_traversal()
            
        #visiting root
        ele.append(self.data)
        
        return ele
    def search(self,val):
        if(self.data==val):
            return True
        
        if(val<self.data):
            if(self.left):
                return self.left.search(val)
            else:
                return False
        else:
            if(self.right):
                return self.right.search(val)
            else:
                return False
    def sum_of_Nodes(self,root):
        if(root is None):
            return 0
        return root.data + self.sum_of_Nodes(root.left)+self.sum_of_Nodes(root.right)
        
def build_tree(num):
    root=Binarysearchtree(num[0])
    for i in range(1,len(num)):
        root.add_child(num[i])
    return root
    
if __name__=='__main__':
    from binarytree import build,bst
    num=[4,2,5,7,1,9,3,15,13,76]
    req_tree=build_tree(num)
    print(req_tree.inorder_traversal())
    print(req_tree.postorder_traversal())
    print(req_tree.preorder_traversal())
    print(req_tree.search(9))
    print("Height of the binary search tree:",req_tree.height(req_tree))
    print("Size of the binary search tree:",req_tree.size())
    print("Sum of elements of the tree:",req_tree.sum_of_Nodes(req_tree))
    tree=build(num)
    print("Simple Binary tree:",tree)
        
        
        
```
